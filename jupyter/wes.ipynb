```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data/past_sample.csv', delimiter=';', index_col='timestamp')
df.index = pd.to_datetime(df.index, unit='s')

df.memory_percent = df.memory_percent.apply(lambda x: round(x * 100, 2))
df.cpu_percent = df.cpu_percent.apply(lambda x: round(x * 100, 2))
df.req_failed = df.req_failed.apply(lambda x: round(min(100, x), 2))

df
```

```python
import matplotlib.pyplot as plt


_, ax = plt.subplots(2, 2, figsize=(24, 12))
ax[0][0].set_title('Requests per second')
ax[0][0].plot(df.request_per_sec)
ax[0][0].grid()

ax[0][1].set_title('Requests failed (%)')
ax[0][1].plot(df.req_failed)
ax[0][1].grid()

ax[1][0].set_title('CPU consumption (%)')
ax[1][0].plot(df.cpu_percent)
ax[1][0].grid()

ax[1][1].set_title('RAM consumption (%)')
ax[1][1].plot(df.memory_percent)
ax[1][1].grid()
```
